# Path Integrals as an Alternative Approach to Option Pricing

**Abstract**

In this work, we propose the use of path integrals as an innovative alternative strategy to conventional models for pricing financial options. We analyze the case of European call options and obtain the Black-Scholes propagator analytically through the application of functional integrals. By convolving the Black-Scholes propagator with the payoff function, we provide a demonstration that offers an alternative perspective to the fundamental asset pricing theorem. Finally, we implement the Metropolis-Hastings algorithm as a numerical method to evaluate the multidimensional integrals inherent to the functional integral formulation and compare the analytical and numerical results.

**Keywords:** Path Integrals, Option Pricing, Black-Scholes Propagator, Functional Integrals, European Call Options, Metropolis-Hastings Algorithm, Numerical Integration

## Introduction

In 1973, Black and Scholes laid the foundations of the classical option pricing theory. In the same year, and independently of Black and Scholes, Merton derived the same formulation. This theory is based on the dynamics of a specific stochastic process—geometric Brownian motion. Under ideal market conditions, without arbitrage, dividends, and with constant volatilities, the Black-Scholes-Merton formula provides an analytical solution for pricing financial options, especially for simple European call options. However, for more complex financial derivatives characterized by early exercise and path-dependency of the underlying asset, the Black-Scholes-Merton formula does not offer analytical solutions. This challenge has been addressed through the development of specialized numerical methods for pricing exotic derivatives with path-dependent features.

In this context, the present work focuses on contributing to the efficient pricing of financial options through the use of the path integral method. This approach, which has its roots in the works of Wiener and Kac in stochastic calculus and Feynman in quantum mechanics, has gained significant importance in finance. It allows for the application of very powerful and robust analytical and numerical techniques. Given the critical role of options in financial markets—with the over-the-counter derivatives gross market value exceeding 15.5 trillion dollars in the first half of 2020—the development of computational models and methods for option pricing is crucial. Despite the well-known limitations of the Black-Scholes-Merton model's assumptions, its widespread use persists largely due to the availability of closed-form solutions. However, more exotic options require numerical approaches such as finite differences, finite elements, or Monte Carlo methods. In this regard, path integrals offer an elegant alternative formulation. Introduced in financial models by Dash and Linetsky, these integrals—based on the earlier work of Feynman in quantum mechanics and Wiener in Brownian motion—have become standard in mathematical finance.

## The Black-Scholes Propagator

Consider the Black-Scholes equation for a European call, where $S$ is the price of the underlying asset, and $K$ and $T$ denote the strike price and expiration time, respectively. Denoting the value of the European call as $C(S,t)$, the Black-Scholes equation can be written as:

$$
\frac{\partial C}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 C}{\partial S^2} + rS \frac{\partial C}{\partial S} - rC = 0
$$

By making the change of variable $x = \ln(S)$, we obtain the following relationships:

$$
\begin{cases}
\displaystyle \frac{\partial C}{\partial S} = \frac{\partial C}{\partial x}\frac{dx}{dS} = \frac{1}{S}\frac{\partial C}{\partial x}, \\[2ex]
\displaystyle \frac{\partial^2 C}{\partial S^2} = \frac{\partial}{\partial S}\left( \frac{1}{S}\frac{\partial C}{\partial x} \right) = -\frac{1}{S^2}\frac{\partial C}{\partial x} + \frac{1}{S^2}\frac{\partial^2 C}{\partial x^2}.
\end{cases}
$$

Substituting these into the Black-Scholes equation, we can rewrite it as:

$$
\begin{split}
& \frac{\partial C}{\partial t} + \frac{1}{2} \sigma^2 S^2 \left( -\frac{1}{S^2}\frac{\partial C}{\partial x} + \frac{1}{S^2}\frac{\partial^2 C}{\partial x^2} \right) + rS\left( \frac{1}{S}\frac{\partial C}{\partial x} \right) - rC = 0, \\
& \Rightarrow \quad \frac{\partial C}{\partial t} + \frac{1}{2} \sigma^2 \frac{\partial^2 C}{\partial x^2} - \left( \frac{1}{2} \sigma^2 - r \right) \frac{\partial C}{\partial x} - rC = 0.
\end{split}
$$

Separating the temporal and spatial parts, we write:

$$
\frac{\partial C}{\partial t} = \hat{H}_{BS} C,
$$

where the Black-Scholes Hamiltonian is defined as

$$
\hat{H}_{BS} := - \frac{1}{2} \sigma^2 \frac{\partial^2}{\partial x^2} + \left( \frac{1}{2} \sigma^2 - r \right) \frac{\partial}{\partial x} + r.
$$

This formulation reveals the connection between option dynamics and quantum systems, as it allows us to interpret the Black-Scholes equation as a time-dependent Schrödinger equation after the substitution $t \rightarrow \frac{-it}{\hbar}$ (a Wick rotation). This provides a common mathematical framework to describe both systems, meaning insights from one area can be translated to the other. In this analogy, the options market is viewed as a system whose states are described by functions $C(x,t)$ (informally, "financial wave functions") and whose evolution is governed by the Black-Scholes equation (informally, "the financial Schrödinger equation"). Accordingly, we have

$$
\begin{split}
C(x,t) = \langle x|C;t \rangle &= \int_{-\infty}^{\infty} \mathrm{d}x_T \, \langle x|e^{-\tau \hat{H}_{BS}}|x_T \rangle \langle x_T|F \rangle \\
&= \int_{-\infty}^{\infty} \mathrm{d}x_T \, K_{BS}(x_T,T|x,t) \, F(x_T),
\end{split}
$$

where

$$
\begin{cases}
\tau := T - t, \quad \text{with } T \text{ as the time to maturity}, \\
F(x_T) := \langle x_T|F \rangle \quad \text{(the payoff function)}, \\
K_{BS}(x_T,T|x,t) := \langle x|e^{-\tau \hat{H}_{BS}}|x_T \rangle.
\end{cases}
$$

Thus, the Black-Scholes propagator can be computed by evaluating the matrix element $\langle x|e^{-\tau \hat{H}_{BS}}|x_T \rangle$. To do this, we introduce the closure relation:

$$
\hat{I} = \frac{1}{\sqrt{2\pi}} \int \mathrm{d}k \, |k\rangle \langle k|,
$$

and since the wave number $k$ is related to the momentum $p$ by $k = \frac{p}{\hbar}$ (with $\hbar = 1$ in this financial context), the propagator becomes:

$$
K_{BS}(x_T,T|x,t) = \int \frac{dp}{\sqrt{2\pi}} \, \langle x_T|e^{-\tau \hat{H}_{BS}}|p \rangle \langle p|x \rangle.
$$

To evaluate this integral, note that from the Black-Scholes Hamiltonian we have

$$
\langle x_T|e^{-\tau \hat{H}_{BS}}|p \rangle = e^{-\tau \hat{H}_{BS}} \langle x_T|p \rangle = e^{-\tau \hat{H}_{BS}} \frac{e^{ipx_T}}{\sqrt{2\pi}}.
$$

We then need to determine how the operator $e^{-\tau \hat{H}_{BS}}$ acts on $e^{ipx_T}$. Using a Taylor expansion of the exponential,

$$
\begin{split}
e^{-\tau \hat{H}_{BS}} e^{ipx_T} &= \left[ \hat{I} - \tau \hat{H}_{BS} + \tau^2 \hat{H}_{BS}^2 + \cdots \right] e^{ipx_T} \\
&= \left\{ \hat{I} - \tau \left[ \frac{\sigma^2 p^2}{2} + i\left( \frac{\sigma^2}{2} - r \right)p + r \right] + \tau^2 \left[ \frac{\sigma^2 p^2}{2} + i\left( \frac{\sigma^2}{2} - r \right)p + r \right]^2 + \cdots \right\} e^{ipx_T} \\
&= e^{-\tau \left[ \frac{\sigma^2 p^2}{2} + i\left( \frac{\sigma^2}{2} - r \right)p + r \right]} e^{ipx_T}.
\end{split}
$$

Substituting this result into the expression for the propagator yields

$$
K_{BS}(x_T,T|x,t) = \int \frac{dp}{2\pi} \, e^{-\tau \left[ \frac{\sigma^2 p^2}{2} + ip\left( \frac{\sigma^2}{2} - r \right) + r \right]} e^{ip(x_T - x)}.
$$

Next, we complete the square in the exponent:

$$
\begin{split}
& -\tau \left[ \frac{\sigma^2 p^2}{2} + ip\left( \frac{\sigma^2}{2} - r \right) + r \right] + ip(x_T - x) \\
& \quad = -\frac{\tau\sigma^2}{2}p^2 + i\left[ x_T - x + \left( r - \frac{\sigma^2}{2} \right)\tau \right]p - r\tau \\
& \quad = -\frac{\tau\sigma^2}{2} \left\{ p - \frac{i}{\tau\sigma^2}\left[ x_T - x + \left( r - \frac{\sigma^2}{2} \right)\tau \right] \right\}^2 - \frac{\left[ x_T - x + \left( r - \frac{\sigma^2}{2} \right)\tau \right]^2}{2\tau\sigma^2} - r\tau.
\end{split}
$$

Finally, evaluating the Gaussian integral, we obtain:

$$
K_{BS}(x_T,T|x,t) = \frac{e^{-r\tau}}{\sqrt{2\pi\tau\sigma^2}} \exp\left[ -\frac{\left( x_T - x + \left( r - \frac{\sigma^2}{2} \right)\tau \right)^2}{2\tau\sigma^2} \right].
$$

## Derivation of the Black-Scholes Propagator via Path Integrals

In what follows we use the path integral method—originally developed in the context of quantum mechanics—to derive the Black-Scholes propagator. It is worth noting that although the propagator is obtained analytically, yielding an expression identical to that derived previously, the path integral method constitutes a disruptive and innovative alternative tool that, combined with robust and efficient numerical techniques, allows the modeling of more complex options that might be impossible to tackle with other methods.

Let us begin by transitioning our problem from the Hamiltonian formulation to the Lagrangian formulation. This means we must perform a Legendre transformation on the classical Hamiltonian by replacing:

$$
\frac{\partial}{\partial x} \quad \rightarrow \quad p,
$$ 

that is:

$$
H_{BS} = - \frac{1}{2}\sigma^2 p^2 + \left(\frac{1}{2}\sigma^2 - r\right)p + r
$$

On the other hand, by using Hamilton's equations we obtain the following relations:

$$
\begin{cases}
\dot{x} = \frac{\partial H_{BS}}{\partial p} = -\sigma^2 p - r + \frac{\sigma^2}{2}, \\[2ex]
-\dot{p} = \frac{\partial H_{BS}}{\partial x} = 0.
\end{cases}
$$

Thus, we have:

$$
\begin{aligned}
L_{BS} & = \dot{x}\, p - H_{BS} \\
& = \left(-\sigma^2 p - r + \frac{\sigma^2}{2}\right)p - \left[- \frac{1}{2}\sigma^2 p^2 + \left(\frac{1}{2}\sigma^2 - r\right)p + r\right] \\
& = -\frac{\sigma^2}{2}p^2 - r.
\end{aligned}
$$

Now, the Black-Scholes Lagrangian takes the following form:

$$
L_{BS}[x(t),\dot{x}(t)] = -\frac{1}{2\sigma^2}\left(\dot{x} + r - \frac{\sigma^2}{2}\right)^2 - r.
$$

For a short time interval $\Delta t$ we have:

$$
\begin{aligned}
\langle x_j|e^{-\Delta t \hat{H}_{BS}}|x_{j-1}\rangle &= \frac{e^{-r\Delta t}}{\sqrt{2\pi\Delta t\,\sigma^2}}
\exp\left[-\frac{1}{2\Delta t\,\sigma^2}\left(x_j - x_{j-1} + \left(r - \frac{\sigma^2}{2}\right)\Delta t\right)^2\right] \\
& = \frac{e^{-r\Delta t}}{\sqrt{2\pi\Delta t\,\sigma^2}}
\exp\left[-\frac{\Delta t}{2\sigma^2}\left(\frac{\Delta x}{\Delta t} + r - \frac{\sigma^2}{2}\right)^2\right] \\
& = \frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}
\exp\left[-r\Delta t - \frac{\Delta t}{2\sigma^2}\left(\frac{\Delta x}{\Delta t} + r - \frac{\sigma^2}{2}\right)^2\right] \\
& = \frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}
\exp\left[\Delta t\, L_{BS}[x_j,x_{j-1},\Delta t]\right].
\end{aligned}
$$

If we apply to the Black-Scholes case the same recipe as in quantum mechanics—namely, subdividing the time interval into $N$ equal parts $\Delta t$ (so that $\tau = N\Delta t$) we obtain:

$$
\begin{aligned}
K_{BS}(x_N,t_N|x_0,t_0) &= \langle x_N|e^{-\Delta t \hat{H}_{BS}}|x_0\rangle \\
& = \lim_{N\rightarrow \infty}\int dx_1 \cdots \int dx_{N-1}\, \langle x_N|e^{-\Delta t \hat{H}_{BS}}|x_{N-1}\rangle \cdots \\
&\quad \times \langle x_1|e^{-\Delta t \hat{H}_{BS}}|x_0\rangle\left[1 + \mathcal{O}\left(N^{-1}\right)\right] \\
& = \lim_{N\rightarrow \infty}\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\right)^N \left(\prod_{j = 1}^{N-1}\int_{-\infty}^{\infty}dx_j\right)
\exp\left[\Delta t\sum_{j=1}^N L_{BS}[x_j,x_{j-1},\Delta t]\right]\left[1 + \mathcal{O}\left(N^{-1}\right)\right] \\
& = \int \mathcal{D}x \; \exp\left[-\int_{t}^{T} dt\, L_{BS}\right] \\
& = \int \mathcal{D}x \; \exp\left[-S_{BS}\right],
\end{aligned}
$$

where the integration measure is given by:

$$
\mathcal{D}x = \lim_{N\rightarrow \infty}\prod_{j=1}^{N-1}dx_j\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\right)^N.
$$

As mentioned earlier, it is possible to compute the Black-Scholes propagator analytically with this approach. With that objective in mind, consider the discretized Black-Scholes action:

$$
\begin{aligned}
S_{BS} &= \Delta t\sum_{j=1}^N L_{BS}[x_j,x_{j-1},\Delta t] \\
& = -\frac{1}{2\Delta t\,\sigma^2}\sum_{j=1}^N\left[x_j - x_{j-1} + \Delta t\left(r - \frac{\sigma^2}{2}\right)\right]^2 - N\Delta t\,r.
\end{aligned}
$$

Performing the change of variables $z_j = x_j - x_{j-1} + \Delta t\left(r - \frac{\sigma^2}{2}\right)$ and noting that $dz_j = dx_j$ for $j = 1,2,\dots,N$, we have:

$$
\begin{aligned}
\int \mathcal{D}x \; e^{S_{BS}} &= \lim_{N\rightarrow \infty}\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\right)^N
\int dz_1 \cdots \int dz_{N-1}\,\exp\left[-r\tau - \frac{1}{2\sigma^2\Delta t}\sum_{j=1}^N z_j^2\right]
\left[1 + \mathcal{O}\left(N^{-1}\right)\right].
\end{aligned}
$$

At this point, note that we have $N$ integration variables but only $N-1$ integrals, so an additional condition on the variables $z_j$ is required. This additional condition is the boundary condition:

$$
x_N = x_T,
$$

or in other words, we must reach $x_N$ at the end. Due to the alternating signs in the definition of $z_j$, the sum over $j = 1,2,\dots,N-1$ leaves the endpoints unmatched. That is:

$$
\begin{aligned}
\lambda &= \sum_{j=1}^N z_j \\
& = x_N - x_0 + N\Delta t\left(r - \frac{\sigma^2}{2}\right) \\
& \equiv x_T - x + \tau\left(r - \frac{\sigma^2}{2}\right).
\end{aligned}
$$

This allows us to accommodate the boundary condition as:

$$
\lambda - \sum_{j=1}^N z_j = 0.
$$

We can incorporate this into the expression using the Dirac delta:

$$
\begin{aligned}
\int \mathcal{D}x \; e^{S_{BS}} &= \lim_{N\rightarrow \infty} e^{-r\tau}\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\right)^N
\prod_{j=1}^N\int_{-\infty}^{\infty}dz_j\,
\exp\left[-\frac{1}{2\sigma^2\Delta t}\sum_{j=1}^N z_j^2\right]
\delta\left(\lambda - \sum_{j=1}^N z_j\right) \\
&\quad \times \left[1 + \mathcal{O}\left(N^{-1}\right)\right].
\end{aligned}
$$

Using the integral representation of the delta function:

$$
\delta(x-y) = \int_{-\infty}^{\infty}\frac{dp}{2\pi}\,e^{ip(x-y)},
$$

we have:

$$
\delta\left(\lambda - \sum_{j=1}^N z_j\right) = \int_{-\infty}^{\infty}\frac{dp}{2\pi}\,e^{ip\left(\lambda - \sum_{j=1}^N z_j\right)}.
$$

We can rewrite the equation as follows:

$$
\begin{aligned}
\int \mathcal{D}x \; e^{S_{BS}} &= \lim_{N\rightarrow \infty} e^{-r\tau}\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\right)^N
\prod_{j=1}^N\int_{-\infty}^{\infty}dz_j\,
\exp\left[-\frac{1}{2\sigma^2\Delta t}\sum_{j=1}^N z_j^2\right] \\
&\quad \times \int_{-\infty}^{\infty}\frac{dp}{2\pi}\,e^{ip\left(\lambda - \sum_{j=1}^N z_j\right)}
\left[1 + \mathcal{O}\left(N^{-1}\right)\right] \\
& = \lim_{N\rightarrow \infty} e^{-r\tau}\int_{-\infty}^{\infty}\frac{dp}{2\pi}\, e^{-ip\lambda}
\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\int_{-\infty}^{\infty}dz_j\,
e^{-\frac{z_j^2}{2\sigma^2\Delta t}+ipz_j}\right)^N
\left[1 + \mathcal{O}\left(N^{-1}\right)\right].
\end{aligned}
$$

Completing the square:

$$
\begin{aligned}
- \frac{1}{2\Delta t\,\sigma^2} z_j^2 + ipz_j &= - \frac{1}{2\Delta t\,\sigma^2}\left[z_j^2 - 2i\Delta t\,\sigma^2 p\, z_j + \left(i\Delta t\,\sigma^2p\right)^2 - \left(i\Delta t\,\sigma^2p\right)^2\right] \\
&= - \frac{1}{2\Delta t\,\sigma^2}\left(z_j - i\Delta t\,\sigma^2p\right)^2 -\frac{\Delta t\,\sigma^2p^2}{2}.
\end{aligned}
$$

Thus, we have:

$$
\begin{aligned}
\int \mathcal{D}x \; e^{S_{BS}} &= \lim_{N\rightarrow \infty} e^{-r\tau}\int_{-\infty}^{\infty}\frac{dp}{2\pi}\, e^{-ip\lambda}
\left(\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\int_{-\infty}^{\infty}dz_j\,
e^{-\frac{z_j^2}{2\sigma^2\Delta t}+ipz_j}\right)^N
\left[1 + \mathcal{O}\left(N^{-1}\right)\right] \\
& = \lim_{N\rightarrow \infty} e^{-r\tau}\int_{-\infty}^{\infty}\frac{dp}{2\pi}\, e^{-ip\lambda}
\left[\frac{1}{\sqrt{2\pi\Delta t\,\sigma^2}}\sqrt{2\pi\Delta t\,\sigma^2}\, e^{-\frac{\Delta t\,\sigma^2p^2}{2}}\right]^N
\left[1 + \mathcal{O}\left(N^{-1}\right)\right] \\
& = \lim_{N\rightarrow \infty} e^{-r\tau}\int_{-\infty}^{\infty}\frac{dp}{2\pi}\, e^{-ip\lambda}\, e^{-\frac{\tau\,\sigma^2p^2}{2}}
\left[1 + \mathcal{O}\left(N^{-1}\right)\right].
\end{aligned}
$$

Completing the square once more, we can express the integral in Gaussian form:

$$
\begin{aligned}
-\frac{\tau\,\sigma^2}{2}p^2 - i\lambda p &= -\frac{\tau\,\sigma^2}{2}\left[p^2 + \frac{2i\lambda}{\tau\,\sigma^2}p + \left(\frac{i\lambda}{\tau\,\sigma^2}\right)^2 - \left(\frac{i\lambda}{\tau\,\sigma^2}\right)^2\right] \\
& = -\frac{\tau\,\sigma^2}{2}\left(p + \frac{i\lambda}{\tau\,\sigma^2}\right)^2 - \frac{\lambda^2}{2\tau\,\sigma^2}.
\end{aligned}
$$

From which we obtain:

$$
\begin{aligned}
\int \mathcal{D}x \; e^{S_{BS}} &= \lim_{N\rightarrow \infty} e^{-r\tau}\int_{-\infty}^{\infty}\frac{dp}{2\pi}\, e^{-ip\lambda}\, e^{-\frac{\tau\,\sigma^2p^2}{2}}
\left[1 + \mathcal{O}\left(N^{-1}\right)\right] \\
& = \lim_{N\rightarrow \infty} e^{-r\tau}\frac{1}{2\pi}\sqrt{\frac{2\pi}{\tau\,\sigma^2}}\, e^{-\frac{\lambda^2}{2\tau\,\sigma^2}}
\left[1 + \mathcal{O}\left(N^{-1}\right)\right].
\end{aligned}
$$

Using the definition for $\lambda$, we obtain:

$$
K_{BS}(x_T,t_T|x,t) = \int \mathcal{D}x \; e^{S_{BS}} = \frac{e^{-r\tau}}{\sqrt{2\pi\tau\,\sigma^2}}
\exp\left[-\frac{\left(x_T - x + \left(r - \frac{\sigma^2}{2}\right)\tau\right)^2}{2\tau\,\sigma^2}\right].
$$

Surprisingly, we recover the same propagator obtained by the other method.

A pertinent observation regarding the final expression is that if we convolve the Black-Scholes propagator, we obtain:

$$
C(x,t) = e^{-r\tau}\int_{-\infty}^{\infty}dx_T\, \frac{1}{\sqrt{2\pi\tau\,\sigma^2}}
\exp\left[-\frac{\left(x_T - x + \left(r - \frac{\sigma^2}{2}\right)\tau\right)^2}{2\tau\,\sigma^2}\right]F(x_T),
$$

identifying

$$
f_{(x,t)}[x_T] := \frac{1}{\sqrt{2\pi\tau\,\sigma^2}}
\exp\left[-\frac{\left(x_T - x + \left(r - \frac{\sigma^2}{2}\right)\tau\right)^2}{2\tau\,\sigma^2}\right]
$$

as the conditional probability distribution given the initial price $x$ at time $t$, we have:

$$
C(x,t) = e^{-r\tau}\,\mathbb{E}_{(x,t)}\left[F(x_T)\right],
$$

where $\mathbb{E}_{(x,t)}\left[F(x_T)\right]$ is the expectation of the payoff function conditioned on the initial price $x$ at time $t$. In this way, in an alternative and elegant manner, we have demonstrated the **Fundamental Theorem of Asset Pricing**.

## Metropolis-Hastings Algorithm

Although the Black–Scholes propagator can be obtained analytically, thus allowing option prices to be determined via its convolution with the payoff function, it is not always possible to derive closed-form expressions for more complex problems. In these cases, path integrals constitute a very appropriate approach when combined with numerical integration methods. In this regard, if we carefully examine expression for $K_{BS}(x_N,t_N|x_0,t_0)$, we notice that the path integral is essentially a multidimensional integral over the intermediate coordinates weighted by the exponential of the classical Black–Scholes action $S_{BS}$. Therefore, the choice of numerical integration method plays a crucial role in obtaining high precision while saving execution time and computational resources.

Among the known numerical integration methods, the Monte Carlo method stands out. Its precision improves as $\frac{1}{\sqrt{n}}$, where $n$ is the number of sample points, even in $d$-dimensional integrals with arbitrary $d$. This is not the case for other methods such as the trapezoidal rule which, although having a better convergence rate in one dimension (approximately $\sim \frac{1}{n^2}$), performs worse in the multidimensional case (approximately $\sim \frac{1}{n^{2/d}}$ for $d > 4$). Even though methods like Gaussian quadrature improve upon the convergence rate of the trapezoidal rule, the Monte Carlo method remains superior for a sufficiently large number of dimensions.

The fundamental idea behind the Monte Carlo method is to approximate a definite integral using random numbers. In other words, one generates a set of $n$ random numbers $x_j$ in the interval of interest, say $[a, b]$, and sums the corresponding function values $f(x_j)$, so that the integral is estimated by

$$
\int_{a}^{b} f(x) \, dx \approx \frac{b-a}{M}\sum_{j=1}^{M} f(x_j),
$$

where $\frac{b-a}{M}$ represents the estimated average interval width. As $M \rightarrow \infty$, we expect to obtain an accurate approximation of the integral. The major advantage of this approach is its simplicity and ease of implementation. However, its main drawback is its slow convergence, especially when approximating integrals of functions that are very small over large regions of the integration domain, as is the case with Gaussian integrals over infinite intervals encountered in path integrals. Therefore, it becomes mandatory to seek more efficient approaches, particularly when dealing with functions exhibiting nonuniform behavior over the integration domain.

One way to tackle this problem is to construct a generator that pays more attention to the regions where the function is large, i.e., that produces random numbers whose distribution approximates the integrand. Such procedures are known as importance sampling. One of the most well-known importance sampling approaches is the acceptance–rejection method. A classic example is throwing darts at a two-dimensional plane with $x$ and $y$ axes, where the function $f(x)$ is drawn as a curve to delineate the target area. If a dart lands below the curve, the corresponding $x$ value is accepted; otherwise, it is discarded. By iterating this process sufficiently, the random numbers $x_j$ will cluster around the values where $f(x)$ is appreciably large. Formally, the acceptance–rejection method is based on generating two random numbers, $x_j$ and $y_j$, where $y_j$ is chosen in the range between the minimum and maximum of $f(x)$, and selecting those $x_j$ for which $y_j < f(x_j)$. The limitation of this strategy is that it can lead to a high rejection rate, especially in regions where $f(x)$ is small, making it inefficient in those areas.

To overcome this problem, it is necessary to use an algorithm that generates new random numbers with an intrinsic preference for remaining close to the values where $f(x)$ is large. One of the most well-known algorithms of this type is the Metropolis–Hastings algorithm. It is based on generating new samples $x_{j+1}$ from the previous sample $x_j$ by adding a random increment of magnitude $\beta \, \hat{P}$ (i.e., constructing a Markov chain), where $\hat{P}$ is drawn from a random number generator that produces values symmetrically distributed around zero (for example, uniformly distributed between $-1$ and $1$). The parameter $\beta$ determines the step size and is vital to avoid becoming trapped in local maxima and to allow the algorithm to explore the entire space of possible values.

$$
\begin{array}{l}
\textbf{Algorithm: Metropolis–Hastings} \\[1mm]
\textbf{Input:} \; f(x) \text{ (target density)},\; n \text{ (number of iterations)},\; \beta \text{ (step parameter)} \\[1mm]
\textbf{Output:} \; \{x_1, x_2, \dots, x_n\} \text{ (sample approximately distributed according to } f(x)\text{)} \\[1mm]
\hline \\[1mm]
1.\quad \text{Initialize } x_0 \text{ randomly} \\[1mm]
2.\quad \textbf{for } j = 0,1,\dots,n-1 \textbf{ do} \\[1mm]
3.\quad \quad \text{Generate } \hat{P} \text{ from a symmetric proposal distribution (centered at 0)} \\[1mm]
4.\quad \quad \text{Compute } y = x_j + \beta\, \hat{P} \\[1mm]
5.\quad \quad \text{Compute } \alpha = \frac{f(y)}{f(x_j)} \\[1mm]
6.\quad \quad \textbf{if } \alpha \geq 1 \textbf{ then} \\[1mm]
7.\quad \quad \quad x_{j+1} = y \\[1mm]
8.\quad \quad \textbf{else} \\[1mm]
9.\quad \quad \quad \text{Generate } u \text{ uniformly in } [0,1] \\[1mm]
10.\quad \quad \quad \textbf{if } \alpha > u \textbf{ then} \\[1mm]
11.\quad \quad \quad \quad x_{j+1} = y \\[1mm]
12.\quad \quad \quad \textbf{else} \\[1mm]
13.\quad \quad \quad \quad x_{j+1} = x_j \\[1mm]
14.\quad \textbf{end} \\[1mm]
\hline
\end{array}
$$

The Metropolis–Hastings procedure is presented in Algorithm above in pseudocode. It is worth noting that one is free to initiate the algorithm from any random value, although it is recommended to perform a significant number of preliminary iterations to avoid becoming trapped in a region due to an unfortunate initial choice.

## `Python` Implementation

In this work, we implemented the Metropolis–Hastings algorithm using `Python`. We presents a comparison between the analytical and numerical results of the Black–Scholes propagator as a function of $x_T$ for a time window of one year, a risk-free rate of $0.03$, volatility of $0.3$, in $5$ dimensions, and with $1,000,000$ iterations.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ==========
# Parameters
# ==========

# Risk-free rate
r = 0.03
# Annual volatility
sigma = 0.3
# Convenience factor (adjusted drift)
rhat = r - 0.5 * sigma**2
# Time window (1 year)
t = 1
# Number of time intervals
N = 4
# Time step
dt = t / N

# ===========================
# Black-Scholes Propagator
# ===========================

def K_BS(xf):
    """
    Compute the Black-Scholes propagator for a given final log-price xf.

    The propagator is defined as:
    
        K_BS(xf) = exp( -r*t - ((xf + t*rhat)**2) / (2*t*sigma**2) ) 
                   / sqrt(2*pi*t*sigma**2)
    
    Parameters
    ----------
    xf : float or numpy.ndarray
        The final log-price (or position) at time t.

    Returns
    -------
    float or numpy.ndarray
        The value of the Black-Scholes propagator at xf.
    """
    return np.exp(-r * t - ((xf + t * rhat)**2) / (2 * t * sigma**2)) / np.sqrt(2 * np.pi * t * sigma**2)
```

### Monte Carlo with Direct Sampling of Uniform Random Numbers

```python
"""
Title: Path Integral Monte Carlo Estimation of the Black-Scholes Propagator

This script defines random paths to compute the path integral for the Black-Scholes propagator.
It then creates histograms of the computed propagator via Monte Carlo and compares the results
to the analytical solution.
"""
# =====================
# Definition of Paths
# =====================

# Number of path samples
Npath = 20000000

# Matrix 'x' storing the paths:
# The first index labels the time points and the second index labels the different paths.
# The paths are initialized uniformly in the interval [-4*sigma, 4*sigma]
x = -4 * sigma + 8 * sigma * np.random.rand(N, Npath)

# The contribution of each segment of each path to S_{BS} is calculated and stored in 'term1'.
# The contribution of the first segment, which starts at x0 = 0, is computed separately and stored in 'term2'.
term1 = np.sum((x[1:, :] - x[:-1, :] + dt * rhat)**2, axis=0)
term2 = (x[0, :] + dt * rhat)**2

# Compute the exponential of the sum of 'term1' and 'term2'
ExpSBS = np.exp(-(term1 + term2) / (2 * dt * sigma**2))

# ===========
# Histograms
# ===========

# Create histogram for the path integral
ix = np.round(x[-1, :] / (sigma / 5))
ixmin, ixmax = int(np.min(ix)), int(np.max(ix))
xx = np.arange(ixmin, ixmax + 1) * sigma / 5
PathIntegral = np.zeros(ixmax - ixmin + 1)

# Populate PathIntegral
for k in range(Npath):
    ipos = int(ix[k]) - ixmin
    PathIntegral[ipos] += ExpSBS[k]

# Normalize the histogram
i0 = (xx[1] - xx[0]) * np.sum(PathIntegral)
PathIntegral = np.exp(-r * t) * PathIntegral / i0

# =====================
# Comparative Plots
# =====================

# Plot histogram of the Monte Carlo path integral estimation versus the analytical result
plt.hist(xx, bins=xx, weights=PathIntegral, color='blue', alpha=0.5,
         label='Uniform Monte Carlo', density=True)
plt.plot(xx, K_BS(xx), 'r--', label='Analytical')
plt.xlim([-1.5, 1.5])
plt.xlabel(r'$x_T$')
plt.ylabel(r'Propagator $K_{BS}(x_T, T|x, t)$')
plt.legend()
plt.show()

# Plot the results using line and marker plot
plt.plot(xx, PathIntegral, 'k*', label='Uniform Monte Carlo')
plt.plot(xx, K_BS(xx), 'r--', label='Analytical')
plt.xlim([-1.5, 1.5])
plt.xlabel(r'$x_T$')
plt.ylabel(r'Propagator $K_{BS}(x_T, T|x, t)$')
plt.legend()
plt.show()
```

### Monte Carlo with Acceptance–Rejection Sampling

```python
"""
Title: Path Integral Monte Carlo Estimation of the Black–Scholes Propagator 
       (Using Acceptance–Rejection Sampling)

This script defines random paths to compute the path integral for the Black–Scholes propagator.
It then creates histograms of the computed propagator via Monte Carlo Acceptance–Rejection sampling
and compares the results to the analytical solution.
"""

# =====================
# Definition of Paths
# =====================

# Number of path samples
Npath = 20000000

# Matrix 'x' storing the paths:
# The first index labels the time points and the second index labels the different paths.
# The paths are initialized uniformly in the interval [-4*sigma, 4*sigma]
x = -4 * sigma + 8 * sigma * np.random.rand(N, Npath)

# The contribution of each segment of each path to S_{BS} is calculated and stored in 'term1'.
# The contribution of the first segment, which starts at x0 = 0, is computed separately and stored in 'term2'.
term1 = np.sum((x[1:, :] - x[:-1, :] + dt * rhat)**2, axis=0)
term2 = (x[0, :] + dt * rhat)**2

# Compute the exponential of the sum of 'term1' and 'term2'
ExpSBS = np.exp(-(term1 + term2) / (2 * dt * sigma**2))

# ===========
# Histograms
# ===========

# Create histogram for the path integral using Acceptance–Rejection sampling.
ix = np.round(x[-1, :] / (sigma / 5))
ixmin, ixmax = int(np.min(ix)), int(np.max(ix))
xx = np.arange(ixmin, ixmax + 1) * sigma / 5
PathIntegral = np.zeros(ixmax - ixmin + 1)

# Populate PathIntegral
for k in range(Npath):
    ipos = int(ix[k]) - ixmin
    PathIntegral[ipos] += ExpSBS[k]

# Normalize the histogram
i0 = (xx[1] - xx[0]) * np.sum(PathIntegral)
PathIntegral = np.exp(-r * t) * PathIntegral / i0

# =====================
# Comparative Plots
# =====================

# Plot histogram of the Monte Carlo Acceptance–Rejection estimation versus the analytical result
plt.hist(xx, bins=xx, weights=PathIntegral, color='blue', alpha=0.5,
         label='Monte Carlo Acceptance–Rejection', density=True)
plt.plot(xx, K_BS(xx), 'r--', label='Analytical')
plt.xlim([-1.5, 1.5])
plt.xlabel(r'$x_T$')
plt.ylabel(r'Propagator $K_{BS}(x_T, T|x, t)$')
plt.legend()
plt.show()

# Plot the results using a marker and line plot
plt.plot(xx, PathIntegral, 'k*', label='Monte Carlo Acceptance–Rejection')
plt.plot(xx, K_BS(xx), 'r--', label='Analytical')
plt.xlim([-1.5, 1.5])
plt.xlabel(r'$x_T$')
plt.ylabel(r'Propagator $K_{BS}(x_T, T|x, t)$')
plt.legend()
plt.show()
```

### Metropolis-Hastings Implementation

```python
"""
Title: Path Integral Monte Carlo Estimation of the Black–Scholes Propagator 
       (Using the Metropolis-Hastings Algorithm)

This script implements the Metropolis-Hastings algorithm to generate random paths,
which are then used to compute the path integral for the Black–Scholes propagator.
The results are compared against the analytical solution.
"""

# ================================
# Metropolis-Hastings Algorithm
# ================================

def MetropolisHastings(h, beta, iterations, x0):
    """
    Implements the Metropolis-Hastings algorithm to generate a Markov chain.
    
    Parameters:
    h (function): Probability density function used for acceptance criterion.
    beta (float): Step size scaling parameter.
    iterations (int): Number of iterations.
    x0 (float): Initial value of the chain.
    
    Returns:
    np.ndarray: Generated Markov chain.
    """
    x = np.zeros(iterations)
    x[0] = x0

    for i in range(1, iterations):
        x_candidate = x[i - 1] + beta * np.random.normal()

        # Acceptance criterion
        alpha = min(1, h(x_candidate) / h(x[i - 1]))

        # Metropolis acceptance-rejection step
        if np.random.uniform() < alpha:
            x[i] = x_candidate
        else:
            x[i] = x[i - 1]

    return x

# ===================
# Path Definition
# ===================

# Number of paths
Npath = 200000

# =============================
# Metropolis-Hastings Arguments
# =============================

HFunc = lambda x: np.exp(-x**2 / (2 * (2 * sigma)**2))  # Target distribution
x0 = 0.0  # Initial condition
beta = np.sqrt(sigma**2 * dt)  # Step size

# Generate Metropolis-Hastings samples
yin = MetropolisHastings(HFunc, beta, 1000000, x0) 
y = MetropolisHastings(HFunc, beta, Npath * N, yin[-1])
x = y.reshape((N, Npath))

# ============================================
# Terms for the Black–Scholes Propagator
# ============================================

# Compute the contributions to the propagator
term1 = np.sum((x[1:, :] - x[:-1, :] + dt * rhat)**2, axis=0)
term2 = (x[0, :] + dt * rhat)**2

# Compute the exponential factor of the propagator
ExpSBS = np.exp(-(term1 + term2) / (2 * dt * sigma**2))

# Create histogram for the path integral
ix = np.round(x[-1, :] / (sigma / 5)).astype(int)
ixmin, ixmax = np.min(ix), np.max(ix)
xx = np.arange(ixmin, ixmax + 1) * sigma / 5
PathIntegral = np.zeros(ixmax - ixmin + 1)

# Populate PathIntegral
for k in range(Npath):
    ipos = ix[k] - ixmin
    PathIntegral[ipos] += ExpSBS[k]

# Normalize the path integral
i0 = (xx[1] - xx[0]) * np.sum(PathIntegral)
PathIntegral = np.exp(-r * t) * PathIntegral / i0

# =====================
# Comparative Plots
# =====================

plt.hist(x[-1, :], bins=xx, weights=ExpSBS, color='blue', alpha=0.5, label='Metropolis-Hastings', density=True)
plt.plot(xx, K_BS(xx), 'r--', label='Analytical')
plt.xlabel(r'$x_T$')
plt.ylabel(r'Propagator $K_{BS}(x_T, T|x, t)$')
plt.legend()
plt.xlim([-1.5, 1.5])
plt.show()

plt.plot(xx, PathIntegral, 'k*', label='Metropolis-Hastings')
plt.plot(xx, K_BS(xx), 'r--', label='Analytical')
plt.xlabel(r'$x_T$')
plt.ylabel(r'Propagator $K_{BS}(x_T, T|x, t)$')
plt.legend()
plt.xlim([-1.5, 1.5])
plt.show()
```

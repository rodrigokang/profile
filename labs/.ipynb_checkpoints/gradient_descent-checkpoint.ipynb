```
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
import scipy as scp
import pandas as pd
import seaborn as sns
```

```
def f(X):
  x, y = X
  f = np.cos(x + y) + np.sin(x) + np.cos(y)
  return f
```

```
f([0,0])
```

```
xlist = np.linspace(3, 8, 200)
ylist = np.linspace(1, 5, 200)
x, y = np.meshgrid(xlist, ylist)
z = np.cos(x + y) + np.sin(x) + np.cos(y)

fig, axes = plt.subplots(figsize=(15, 12), subplot_kw={'projection': '3d'})
ax = fig.gca(projection = "3d")
ax.set_xlabel("x", fontsize=20, color="blue")
ax.set_ylabel("y", fontsize=20, color="blue")
ax.set_zlabel("f(x,y)", fontsize=20, color="blue")
ax.plot_surface(x, y, z, cmap=mpl.cm.Blues)
plt.show()
```

```
xlist = np.linspace(3, 7, 100)
ylist = np.linspace(3, 5, 100)
x, y = np.meshgrid(xlist, ylist)
z = np.cos(x + y) + np.sin(x) + np.cos(y)
fig, ax = plt.subplots(figsize=(15, 8))
cp = ax.contourf(x, y, z)
fig.colorbar(cp)
ax.set_title('Curvas de nivel', fontsize=20, color="blue")
ax.set_xlabel("x", fontsize=20, color="blue")
ax.set_ylabel("y", fontsize=20, color="blue")
plt.show()
```

```
def grad_f(X):
  x, y = X
  D_xf = np.cos(x) - np.sin(x + y)
  D_yf = - np.sin(y) - np.sin(x + y)
  return np.array([D_xf,D_yf])
```

```
grad_f([0,0])
```

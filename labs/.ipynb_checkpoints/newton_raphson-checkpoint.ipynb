```python
def Newton(f,Df,x0,epsilon,max_iter):
  xn = x0
  for n in range(0,max_iter):
    fxn = f(xn)
    if abs(fxn) < epsilon:
      print('Solución encontrada luego de',n,
      'iteraciones.')
      return xn
    Dfxn = Df(xn)
    if Dfxn == 0:
      print('Derivada cero. No se encontró ninguna solución.')
      return None
    xn = xn - fxn/Dfxn
  print('Se superaron las iteraciones máximas. No se encontró solución.')
  return None
```

Ejemplos

```python
f = lambda x: x**2 - x - 1
Df = lambda x: 2*x - 1

Newton(f,Df,1,1e-8,10)
```

```python
def Secante(f,a,b,N):
    if f(a)*f(b) >= 0:
        print("El método de la secante falla.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        m_n = a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
        f_m_n = f(m_n)
        if f(a_n)*f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n)*f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Solución exacta encontrada.")
            return m_n
        else:
            print("El método falla.")
            return None
    return a_n - f(a_n)*(b_n - a_n)/(f(b_n) - f(a_n))
```

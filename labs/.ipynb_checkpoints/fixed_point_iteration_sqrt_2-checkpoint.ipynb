# Approximation the Square Root of 2 Using Fixed Point Iteration Method

## Fixed Point Iteration Algorithm

### Pseudocode

$$
\begin{array}{l}
\textbf{Algorithm: Fixed-Point Iteration} \\ 
\textbf{Input:} \, g, x_0, \varepsilon, N \\ 
\textbf{Output:}\, \text{Approximate solution $x^*$ or error message} \\ 
\hline
1. \quad x_n = x_0 \\
2. \quad \textbf{for } n = 0,1,\dots,N-1 \textbf{ do} \\
3. \quad \quad g_x = g(x_n) \\
4. \quad \quad \textbf{if } |g_x - x_n| < \varepsilon \textbf{ then} \\
5. \quad \quad \quad \text{print "Solution found after } n \text{ iterations"} \\
6. \quad \quad \quad \textbf{return } g_x \\
7. \quad \quad x_n = g_x \\
8. \quad \text{print "Maximum iterations exceeded. No solution found"} \\
9. \quad \textbf{return } \text{None} \\
\hline
\end{array}
$$

### `python` implementation

#### Code without numerical libraries of `python`

```python

def fixed_point_iteration(g, x0, epsilon, max_iter):
    """
    Performs fixed-point iteration to find a solution to x = g(x).
    
    Parameters:
    g (function): The function to iterate.
    x0 (float): Initial guess.
    epsilon (float): Convergence tolerance.
    max_iter (int): Maximum number of iterations.

    Returns:
    float: The approximate solution if found within tolerance.
    None: If no solution is found within max_iter.
    """
    xn = x0
    for n in range(max_iter):
        gxn = g(xn)
        
        if abs(gxn - xn) < epsilon:
            print(f"Solution found after {n} iterations.")
            return xn
        
        xn = gxn  # Update xn for next iteration

    print("Maximum iterations exceeded. No solution found.")
    return None
```

## Problem

Using the fixed point iteration method, approximate $\sqrt{2}$ with an accuracy of $10^{-4}$. Track the progress of the approximation at each step.

## Solution

````python
def fixed_point_iteration(g, x0, epsilon, max_iter):
    """
    Fixed-Point Iteration Method to approximate a solution.

    Parameters:
    g (function): The iteration function g(x).
    x0 (float): Initial guess.
    epsilon (float): Tolerance for stopping condition.
    max_iter (int): Maximum number of iterations.

    Returns:
    float: Approximated root or None if it fails.
    """
    xn = x0
    for n in range(max_iter):
        gxn = g(xn)
        print(f"Iteration {n+1}: x_n = {xn:.6f}")

        if abs(gxn - xn) < epsilon:
            print(f"Solution found after {n+1} iterations: x = {gxn:.6f}")
            return gxn
        
        xn = gxn

    print("Maximum iterations exceeded. No solution found.")
    return None

# Define the iteration function g(x) for sqrt(2)
def g(x):
    return 0.5 * (x + 2 / x)  # Derived from x = sqrt(2) -> x = (x + 2/x) / 2

# Initial guess
x0 = 1.5
epsilon = 1e-4
max_iter = 100

# Run fixed-point iteration
fixed_point_iteration(g, x0, epsilon, max_iter)
````

````python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Function for g(x) as defined for sqrt(2)
def g(x):
    return 0.5 * (x + 2 / x)  # Derived from x = sqrt(2) -> x = (x + 2/x) / 2

# Fixed-point iteration function with tracking and plotting
def fixed_point_iteration_with_plot(g, x0, epsilon, max_iter):
    xn = x0
    iterations = [xn]
    
    for n in range(max_iter):
        gxn = g(xn)
        iterations.append(gxn)
        
        if abs(gxn - xn) < epsilon:
            break
        
        xn = gxn

    # Plotting the function g(x), y=x, and the iterations
    x_vals = np.linspace(1, 2, 400)
    y_gx = g(x_vals)
    
    plt.plot(x_vals, y_gx, label=r'$g(x) = \frac{x + \frac{2}{x}}{2}$', color='blue')
    plt.plot(x_vals, x_vals, label=r'$y = x$', color='red', linestyle='--')
    
    # Plot the iterations
    plt.plot(iterations, iterations, marker='o', color='green', label='Approximations')
    
    plt.title(r'Fixed Point Iteration for Approximating $\sqrt{2}$')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

    return iterations[-1]

# Initial guess and parameters
x0 = 1.5
epsilon = 1e-4
max_iter = 100

# Run the function and plot the results
fixed_point_iteration_with_plot(g, x0, epsilon, max_iter)
````

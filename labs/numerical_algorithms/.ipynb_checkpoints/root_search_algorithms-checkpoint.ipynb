# Algoritmos de búsqueda de raíces

## Aproximación de $\sqrt{2}$

En esta sección exploraremos distintos algoritmos numéricos para la búsqueda de raíces de funciones reales. Como caso de estudio, utilizaremos un ejemplo clásico y sencillo: la aproximación de la raíz cuadrada de 2. Este problema, además de ser históricamente relevante, permite ilustrar con claridad los principios y particularidades de cada método.

A lo largo del análisis, veremos cómo distintos enfoques —como el método de bisección, el método de Newton-Raphson, el método de la secante, entre otros— pueden aplicarse para encontrar soluciones aproximadas a la ecuación:

$$
f(x) = x^2 - 2 = 0
$$

Estudiaremos la lógica detrás de cada algoritmo, su velocidad de convergencia, condiciones de aplicabilidad y posibles ventajas y desventajas. Este enfoque nos permitirá comprender en profundidad cómo funcionan estas herramientas fundamentales del análisis numérico.

### Algoritmo de las Bisecciones

En este apartado, aplicaremos el **método de bisección** para estimar la raíz cuadrada de 2, resolviendo la ecuación no lineal:

$$
f(x) = x^2 - 2 = 0
$$

Nuestro objetivo es determinar cuántas iteraciones se requieren para aproximar la solución con una **precisión de $10^{-4}$**. Para ello, en cada iteración registraremos el progreso del método, observando cómo los extremos del intervalo se van ajustando y cómo evoluciona la estimación de la raíz.

Los resultados se presentarán en una tabla que incluye las siguientes columnas:

- El número de iteración $N$,
- Los extremos actualizados del intervalo $[a_N, b_N]$,
- El punto medio del intervalo $p_N = \frac{a_N + b_N}{2}$,
- La aproximación obtenida de $\sqrt{2}$ en esa iteración, es decir, el valor de $p_N$.

Esta tabla nos permitirá visualizar de forma clara la convergencia del algoritmo y analizar su comportamiento iterativo paso a paso.

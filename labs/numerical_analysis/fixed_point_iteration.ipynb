```
import numpy as np
import scipy as scp
import scipy.linalg
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

```
def Fixed_point_iteration(g,x0,epsilon,max_iter):
  # epsilson := tolerancia
  xn = x0
  for n in range(0,max_iter):
    gxn = g(xn)
    if abs(gxn - xn) < epsilon:
      print ('Solución encontrada luego de',n,
      'iteraciones.')
      return xn
    xn = gxn
  print('Se superaron las iteraciones máximas. No se encontró solución.')
  return None
  
# Ejemplo

g = lambda x: (1/2) * np.sqrt(10 - x**3)

Fixed_point_iteration(g,1.5,1e-10,35)
```

```
import numpy as np
import scipy as scp
import scipy.linalg
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

```
def Newton(f,Df,x0,epsilon,max_iter):
  xn = x0
  for n in range(0,max_iter):
    fxn = f(xn)
    if abs(fxn) < epsilon:
      print('Solución encontrada luego de',n,
      'iteraciones.')
      return xn
    Dfxn = Df(xn)
    if Dfxn == 0:
      print('Derivada cero. No se encontró ninguna solución.')
      return None
    xn = xn - fxn/Dfxn
  print('Se superaron las iteraciones máximas. No se encontró solución.')
  return None
```

Ejemplos

```
f = lambda x: x**2 - x - 1
Df = lambda x: 2*x - 1

Newton(f,Df,1,1e-8,10)
```

```
f = lambda x: x**3 + 4*x**2 - 10
Df = lambda x: 3*x**2 + 8*x

Newton(f,Df,1,1e-8,10)
```

```
p = lambda x: x**3 - x**2 - 1
Dp = lambda x: 3*x**2 - 2*x

Newton(p,Dp,1,1e-10,10)
```

```
f = lambda x: x**(1/3)
Df = lambda x: (1/3)*x**(-2/3)

Newton(f,Df,0.1,1e-2,100)
```

# Approximating $\sqrt{2}$ Using the Bisection Algorithm

## Problem

Find an approximation to $\sqrt{2}$ correct to within $10^{−4}$ using the bisection algorithm. 

Hint: Consider $f(x) = x^2 − 2$

## Solution

bisection_pseudocode.PNG

```python
import numpy as np
import scipy as scp
import scipy.linalg
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
def Biseccion(f,a,b,N):
    if f(a)*f(b) >= 0:
        print("El método de bisección falla.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        p_n = (a_n + b_n)/2
        f_p_n = f(p_n)
        if f(a_n)*f_p_n < 0:
            a_n = a_n
            b_n = p_n
        elif f(b_n)*f_p_n < 0:
            a_n = p_n
            b_n = b_n
        elif f_p_n == 0:
            print("Solución exacta encontrada")
            return p_n
        else:
            print("El método de bisección falla.")
            return None
    return (a_n + b_n)/2
```

```python
#Error

def Funcion_Error(a,b,N):
    return (b - a) / 2**(N+1)
```

```python
# Ejemplo 1
    
f = lambda x: x**3 + 4*x**2 - 10
Sol = Biseccion(f,1,2,13)
print ("Solución aproximada =", Sol)

Error = Funcion_Error(1,2,13)
print("El error cometido es =", Error)
```

```python
fig = plt.figure(figsize=(8, 2.5), dpi=100)

left, bottom, width, height = 0.1, 0.1, 1, 2
ax = fig.add_axes((left, bottom, width, height))

x = np.linspace(-1,3,100)
y = x**3 + 4*x**2 - 10
ax.plot(x, y, color="blue",  linestyle='-.', label="f(x)")
ax.set_xlabel("x", labelpad=15, fontsize=18, color="blue")
ax.set_ylabel("f(x)", labelpad=15, fontsize=18, 
color="blue")
ax.set_title("Método de las bisecciones", fontsize=16, 
color="blue")
ax.legend()
plt.grid()
```

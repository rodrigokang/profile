# Birthday Problem

## Problem

Determine the minimum number of people, $M$, needed in a room such that the probability of at least two individuals sharing the same birthday exceeds $1/2$. Assume that every day of the year is equally likely to be a birthday and disregard leap years, considering a year to have $N = 365$ days. Solve this both analytically and using code.

## Solution

Let $A$ be the event "all birthdays are different"; therefore, its complement, $A'$, represents the event "at least two individuals share the same birthday".

Assuming there are exactly $365$ days in a year and that all birthdays are equally likely, although in reality this ignores leap years, cultural and seasonal variations in birth rates, and certain scheduling preferences for births.

* $p_1$: Probability that the first of the $M$ people has any birthday $\Rightarrow p_1 = 365/365 = 1$.
* $p_2$: Probability that the second person has a birthday different from the first $\Rightarrow p_2 = 364/365$.
* $p_3$: Probability that the third person has a birthday different from the first two $\Rightarrow p_3 = 363/365$.
* $\cdots$
* $p_M$: Probability that the $M$-th person has a birthday different from the others $\Rightarrow p_M = (365 - M + 1)/(365)$.

Thus, the probability that all birthdays are different is:

$$
\begin{split}
P(A) & = p_1 \cdot p_2 \cdot p_3 \cdots p_M \\
     & = \frac{365}{365} \cdot \frac{364}{365} \cdot \frac{363}{365} \cdots \frac{365 - M + 1}{365} \\
     & = 1 \cdot \frac{365 - 1}{365} \cdot \frac{365 - 2}{365} \cdots \frac{365 - M + 1}{365} \\
     & = \left(1 - \frac{1}{365} \right) \cdot \left(1 - \frac{2}{365}\right) \cdots \left(1 - \frac{M - 1}{365}\right) \\
\end{split}
$$

Since our goal is to find the minimum number of people such that the probability of at least two of them sharing the same birthday exceeds $1/2$, and given that:  

$$
P(A') = 1 - P(A),
$$  

the problem is solved in the same way by imposing $P(A) < 1/2$. Therefore, Let $P(A)$ be denoted by $p$, and by taking natural logarithms, we obtain:

$$
\ln p = \ln \left(1 - \frac{1}{365} \right) + \ln \left(1 - \frac{2}{365}\right) + \cdots + \ln \left(1 - \frac{M - 1}{365}\right)
$$

From calculus, we know that:

$$
\ln(1 - x) = - x - \frac{x^2}{2} - \frac{x^3}{3} - \cdots
$$

Thus, the equation for $\ln p$ can be approximated as:

$$
\begin{split}
\ln(1 - x) & = - \frac{1}{365} - \frac{1}{2}\frac{1}{365^2} - \cdots - 
\frac{2}{365} - \frac{1}{2}\frac{2^2}{365^2} - \cdots - 
\frac{M-1}{365} - \frac{1}{2}\frac{(M - 1)^2}{365^2} - \cdots \\
           & = - \left[ \frac{1 + 2 + \cdots + (M-1)}{365} \right] - \frac{1}{2} \left[ \frac{1^2 + 2^2 + \cdots + (M-1)^2}{365^2} \right] - \cdots \\
           & = - \frac{M(M-1)}{2(365)} - \frac{M(M-1)(2M-1)}{12(365^2)} - \cdots
\end{split}
$$

For $M$ much smaller than $365$, let say $M < 30$, the second and higher-order terms on the right-hand side become negligible compared to the first term. Therefore, a good approximation in this case is:

$$
\ln p = - \frac{M(M-1)}{730}
$$

For $p=1/2$, $\ln p = - \ln 2 \approx - 0.693$ we have:

$$
\begin{split}
\frac{M(M-1)}{730} = 0.693 \,\,\, \Rightarrow \,\,\, (M-23)(M+22) = 0
\end{split}
$$

Thus, we conclude that when $M$ exceeds 23, the probability of at least two people sharing the same birthday becomes greater than $50\%$.

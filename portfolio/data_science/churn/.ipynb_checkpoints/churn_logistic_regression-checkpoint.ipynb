# Churn Prediction in Non-Contractual Relationships Based on RFM Analysis for the Retail Market

**Abstract**  

This project aims to predict customer churn in non-contractual relationships within the retail market using an **RFM**-based approach. To achieve this, the **Recency** (**R**), **Frequency** (**F**), and **Monetary Value** (**M**) metrics were calculated from the customers' purchase history. These metrics were then transformed using the **Box-Cox** transformation and normalized with **Z-Scoring** to obtain their scaled versions: $S_R$, $S_F$, and $S_M$. Using these values, an RFM score $S_{RFM}$ was constructed as a weighted combination of $S_R$, $S_F$, and $S_M$. The weights assigned to each component were determined based on their explained variances, ensuring that each metric's contribution reflected its relative importance in the data distribution.

Since customer relationships are non-contractual, churn was defined using a **K-Means** segmentation approach. A cluster labeled **"Low-Activity Customers"** was identified, and a customer is considered a churner if the following conditions are met:

1. They belong to the "Low-Activity Customers" cluster.
2. Their RFM score $S_{RFM}$ is lower than the cluster's average.
3. Their recency is greater than their average latency, where latency is defined as the time elapsed between transactions.

Finally, $S_R$, $S_F$, $S_M$, and $S_{RFM}$ were used as predictors in a logistic regression model to classify customers into two groups: those who churn and those who remain active. This approach provides an analytical tool to identify high-risk churn customers and enables the design of more effective retention strategies based on transactional behavior.

**Keywords:** *RFM analysis*, *churn prediction*, *K-Means segmentation*, *low-activity customers*, *recency*, *frequency*, *monetary value*, *latency*, *Box-Cox transformation*, *Z-Scoring*, *logistic regression*, *retail market*.

## Design and Implementation of an RFM-based Scoring System

### Strategy to Address the Skewness of Variables R, F, and M

In order to carry out the clustering process by implementing the `K-Means` algorithm, we need to preprocess the data to meet the assumptions of no skewness. To address the skewness in the variables of recency (R), frequency (F), and monetary value (M), we apply the Box-Cox transformation. The Box-Cox transformation is a family of potential techniques widely used in statistics. These transformations are used to mitigate skewness in distributions, counteract unequal variances based on the predictor variable values, and, fundamentally, to mitigate non-linearity in the relationships between variables, thereby improving the correlation between them. This approach is named after statisticians **George E. P.** Box and **David Cox**.

In our project, the transformation is performed using the `boxcox` function from the [`scipy.stats`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox.html) library. The formula for the Box-Cox transformation is given by:

$$
y(\lambda) = \left\{
  \begin{array}{ll}
    \frac{{x^\lambda - 1}}{{\lambda}}, & \text{if } \lambda \neq 0 \\
    \log(x), & \text{if } \lambda = 0
  \end{array}
\right.
$$


````python
# Box-Cox Transformation

# Apply Box-Cox to 'recency'
input_data['recency'], LambdaRecencia = boxcox(input_data['recency'] + 1)
# Apply Box-Cox to 'frequency'
input_data['frequency'], LambdaFrecuencia = boxcox(input_data['frequency'] + 1)
# Apply Box-Cox to 'monetary value'
input_data['monetary'], LambdaMonto = boxcox(input_data['monetary'] + 1)

input_data.head()
````

### Z-Scoring (Standardization of Variables)

We must also standardize the transformed variables to ensure that the data is normalized for the `K-Means` algorithm, which uses Euclidean distance in its clustering process. This preprocessing enhances the model's effectiveness by ensuring that the variables are comparable and contribute equally to the formation of the clusters. The standardization of the variables is performed as follows:

$$
\left\{
\begin{array}{c}
S_R = \frac{R - \mu_R}{\sigma_R} \\
S_F = \frac{F - \mu_F}{\sigma_F} \\
S_M = \frac{M - \mu_M}{\sigma_M}
\end{array}
\right.
$$

````python
# Standardization for recency
input_data['s_r'] = ((input_data['recency'] - input_data['recency'].mean()) / 
                      input_data['recency'].std())

# Standardization for frequency
input_data['s_f'] = ((input_data['frequency'] - input_data['frequency'].mean()) /
                     input_data['frequency'].std())

# Standardization for monetary value
input_data['s_m'] = ((input_data['monetary'] - input_data['monetary'].mean()) / 
                      input_data['monetary'].std())
````

### Optimal Number of Clusters - Inertia (Elbow Method) 

The determination of the optimal number of clusters is a crucial aspect of implementing the `K-Means` algorithm, as it is a key hyperparameter of the model. To address this challenge, we adopted the **elbow method** as our primary approach.

The elbow method involves plotting **inertia** against the number of clusters and selecting the point where there is no significant reduction in inertia, indicating the optimal number of clusters. This technique provides a balance between minimizing within-cluster variance and avoiding excessive fragmentation.

Inertia, represented by the sum of squared distances between each point and the center of its assigned cluster, was the first metric evaluated using the Elbow Method. This method involves plotting inertia as a function of the number of clusters and looking for the "elbow" in the curve, which indicates the point where adding more clusters no longer provides a significant improvement in the reduction of inertia.

By applying this method, we ensured an effective segmentation of customers in our RFM dataset, selecting the most appropriate number of clusters based on the trade-off between model complexity and interpretability.

````python
# Extract the relevant columns for clustering
data = input_data[['s_r', 's_f', 's_m']]

# Initialize list to store inertia results
inertia_values = []

# Calculate the sum of squared distances (inertia) for different k values
for k in range(2, 11):  # Try k from 2 to 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    
    inertia = kmeans.inertia_
    inertia_values.append(inertia)

# Create a figure for the Elbow Method
plt.figure(figsize=(8, 5))

# Elbow method
plt.plot(range(2, 11), inertia_values, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.title('Elbow Method')
plt.grid(True)

plt.show()
````

### RFM score

The final RFM Score is calculated as a weighted combination of the individual scores of Recency ($S_R$), Frequency ($S_F$), and Monetary ($S_M$). The formula for the RFM Score ($S_{RFM}$) is expressed mathematically as:

$$
S_{RFM} = w_{S_R} \cdot S_R + w_{S_F} \cdot S_F + w_{S_M} \cdot S_M
$$

Here, $w_{S_R}$, $w_{S_F}$, $w_{S_M}$ are the weights assigned to each metric, and they are subject to the constraint:

$$
w_{S_R} + w_{S_F} + w_{S_M} = 1
$$

These weights $(w_{S_R}, w_{S_F}, w_{S_M})$ are determined using the explained variances of each metric:

$$
\left\{
\begin{array}{c}
w_{S_R} = \frac{\sigma_{S_R}^2}{\sigma_{S_R}^2 + \sigma_{S_F}^2 + \sigma_{S_M}^2} \\
w_{S_F} = \frac{\sigma_{S_F}^2}{\sigma_{S_R}^2 + \sigma_{S_F}^2 + \sigma_{S_M}^2} \\
w_{S_M} = \frac{\sigma_{S_M}^2}{\sigma_{S_R}^2 + \sigma_{S_F}^2 + \sigma_{S_M}^2}
\end{array}
\right.
$$

This formulation allows generalization to $N$ variables, meaning the RFM Score ($S_{RFM}$) can be adapted to incorporate a broader set of metrics ($S_i$). The general equation is defined as:

$$
S_{RFM} = \sum_{i=1}^N w_{S_i} \cdot S_{i}
$$

where $w_{S_i}$ represents the weights associated with each individual metric, and the additional constraint is:

$$
\sum_{i=1}^N w_{S_i} = 1
$$

where

$$
w_{S_i} = \frac{\sigma_{S_i}^2}{\sum_{i=1}^N\sigma_{S_i}^2}
$$

This flexible approach allows effective customization of the RFM Score according to the specific needs of the analysis, enabling the inclusion of a variable number of metrics to improve or adjust customer categorization.

The definition of scores in each direction $(S_R, S_F, S_M)$ opens the possibility to classify the resulting clusters, which can be summarized in the table below. This table provides a valuable reference framework for interpreting customer profiles in each category. The upward arrows () indicate that the score in the corresponding direction for that customer type is higher than the overall average of the dataset, while the downward arrows () indicate that the metric is lower than the overall average. This information is crucial for strategic decision-making, as it allows the identification of specific patterns and trends in customer behavior. For example, "**High Contribution Customers**" are characterized by a Recency ($\overline{S_R}$) below average, but with higher Frequency ($\overline{S_F}$) and Monetary ($\overline{S_M}$), suggesting the importance of maintaining their engagement with the business. This classification provides valuable insights for designing personalized strategies aimed at retention, development, or acquisition of customers based on their specific profiles.

**Cluster classification criterion where $\uparrow$ means higher than the overall average, and $\downarrow$ means lower than the overall average of the dataset.**

````python
variables = ['s_r', 's_f', 's_m']

# List to store the variances for each cluster (corresponding to the weights)
variances_by_cluster = []

# Iterate through each cluster
for cluster in input_data['cluster'].unique():
    cluster_data = input_data[input_data['cluster'] == cluster][variables]
    cluster_variances = cluster_data.var()
    variances_by_cluster.append(cluster_variances)

# Create a DataFrame with the variances for each cluster
variance_df = pd.DataFrame(variances_by_cluster, index=input_data['cluster'].unique(), columns=variables)

# Normalize the variances by cluster so that they sum to 1
normalized_variances = variance_df.div(variance_df.sum(axis=1), axis=0)

var_r = input_data['s_r'].std()**2
var_f = input_data['s_f'].std()**2
var_m = input_data['s_m'].std()**2

# Print the explained variances
print('<<<<<<<<<<<<<<<<<<<<<<<< * >>>>>>>>>>>>>>>>>>>>>>>')

print("Explained variances by cluster:")
print("--------------------------------------------------")
print(normalized_variances)
print("==================================================")
print("Total variances (considering all as a single cluster)")
print("--------------------------------------------------")
print('VarNorm(R) = ', var_r/(var_r + var_f + var_m))
print('VarNorm(F) = ', var_f/(var_r + var_f + var_m))
print('VarNorm(M) = ', var_m/(var_r + var_f + var_m))
print('<<<<<<<<<<<<<<<<<<<<<<<< * >>>>>>>>>>>>>>>>>>>>>>>')
````

````python
# Add the clustering results to the original DataFrame
input_data['cluster'] = cluster_labels

# Add the weights
for cluster in input_data['cluster'].unique():
    mask = input_data['cluster'] == cluster
    # Add the weight columns to each customer in the cluster
    input_data.loc[mask, '$\omega_R$'] = normalized_variances.loc[cluster, 's_r']
    input_data.loc[mask, '$\omega_F$'] = normalized_variances.loc[cluster, 's_f']
    input_data.loc[mask, '$\omega_M$'] = normalized_variances.loc[cluster, 's_m']

# Insert the original columns at the beginning of the DataFrame
input_data.insert(1, 'R', input_data_copy['recency'])
input_data.insert(2, 'F', input_data_copy['frequency'])
input_data.insert(3, 'M', input_data_copy['monetary'])
input_data.insert(4, 'AL', input_data_copy['average_latency'])

# Add the S_{RFM} column to the DataFrame
input_data['s_rfm'] = 0.0
for cluster in input_data['cluster'].unique():
    mask = input_data['cluster'] == cluster
    input_data.loc[mask, 's_rfm'] = (
        - input_data.loc[mask, 's_r'] * input_data.loc[mask, '$\omega_R$'] +
          input_data.loc[mask, 's_f'] * input_data.loc[mask, '$\omega_F$'] +
          input_data.loc[mask, 's_m'] * input_data.loc[mask, '$\omega_M$']
    )

columns_order = ['Customer ID', 'R', 'F', 'M', 'AL',
                 'cluster', '$\omega_R$', '$\omega_F$', '$\omega_M$',
                 's_r', 's_f', 's_m', 's_rfm']

# Reorder the columns in the DataFrame
input_data = input_data.reindex(columns=columns_order)

# Print the DataFrame
input_data.head()
````

````python
# Calculate the mean values of S_R, S_F, and S_M by cluster
means_by_cluster = input_data.groupby('cluster').agg({'s_r':'mean', 's_f':'mean', 's_m':'mean'})

# Calculate the total means of S_R, S_F, and S_M
MeanSr = input_data['s_r'].mean()
MeanSf = input_data['s_f'].mean()
MeanSm = input_data['s_m'].mean()

# Define a function to assign the category
def assign_category(row):
    """
    Assigns a category to a customer based on their RFM values and the cluster and total means.
    
    Args:
        row (pd.Series): Row from the DataFrame with RFM values and the assigned cluster.
        
    Returns:
        str: Category assigned to the customer.
    """
    ClusterMeans = means_by_cluster.loc[row['cluster']]  # Get the means for the corresponding cluster
    if (ClusterMeans['s_r'] < MeanSr and 
        ClusterMeans['s_f'] > MeanSf and 
        ClusterMeans['s_m'] > MeanSm):
        return 'high contribution customers'
    elif (ClusterMeans['s_r'] > MeanSr and 
          ClusterMeans['s_f'] > MeanSf and 
          ClusterMeans['s_m'] > MeanSm):
        return 'important customers to retain'
    elif (ClusterMeans['s_r'] < MeanSr and 
          ClusterMeans['s_f'] < MeanSf and 
          ClusterMeans['s_m'] > MeanSm):
        return 'important customers for development'
    elif (ClusterMeans['s_r'] > MeanSr and 
          ClusterMeans['s_f'] < MeanSf and 
          ClusterMeans['s_m'] > MeanSm):
        return 'important customers for retention'
    elif (ClusterMeans['s_r'] < MeanSr and 
          ClusterMeans['s_f'] > MeanSf and 
          ClusterMeans['s_m'] < MeanSm):
        return 'potential customers'
    elif (ClusterMeans['s_r'] < MeanSr and 
          ClusterMeans['s_f'] < MeanSf and 
          ClusterMeans['s_m'] < MeanSm):
        return 'recent customers'
    elif (ClusterMeans['s_r'] > MeanSr and 
          ClusterMeans['s_f'] > MeanSf and 
          ClusterMeans['s_m'] < MeanSm):
        return 'general maintenance customers'
    elif (ClusterMeans['s_r'] > MeanSr and 
          ClusterMeans['s_f'] < MeanSf and 
          ClusterMeans['s_m'] < MeanSm):
        return 'low activity customers'

# Apply the function to assign the category
input_data['category'] = input_data.apply(assign_category, axis=1)
````

````python
# Function to assign values to the 'description' column
def assign_description(category):
    """
    Assigns a description based on the given category.

    Parameters:
    - category (str): The category for which the description will be assigned.

    Returns:
    - str: The description assigned based on the category.
    """
    
    if category == 'high contribution customers':
        return 'r  f  m '
    elif category == 'important customers to retain':
        return 'r  f  m '
    elif category == 'important customers for development':
        return 'r  f  m '
    elif category == 'important customers for retention':
        return 'r  f  m '
    elif category == 'potential customers':
        return 'r  f  m '
    elif category == 'recent customers':
        return 'r  f  m '
    elif category == 'general maintenance customers':
        return 'r  f  m '
    elif category == 'low activity customers':
        return 'r  f  m '
    else:
        return ''  # Handle unspecified categories

# Apply the function to the 'category' column to create the new 'description' column
input_data['description'] = input_data['category'].apply(assign_description)
````

## Churn Prediction

### Churn Definition

````python
# Create the delta_compra column
input_data['delta'] = input_data['R'] - input_data['AL']

# Create the churn column
input_data['real_churn'] = ((input_data['category'] == 'low activity customers') & 
                            (input_data['delta'] > 0) &
                            (input_data['s_rfm'] < input_data[input_data['category'] == 
                             'low activity customers']['s_rfm'].mean())).astype(int)
````

### Train the Churn Model

````python
# Define the parameters
test_size = 0.3  # Training sample size
sampling_strategy = 1.0  # Sampling strategy
n_iter = 30  # Number of combinations to try in RandomizedSearchCV
cv_folds = 5  # Number of folds for cross-validation
random_state = 42  # Set seed for reproducibility

# Set the seed for randomness
np.random.seed(random_state)

# Split the dataset into training and testing sets
X = input_data[['s_r', 's_f', 's_m', 's_rfm_norm']]
y = input_data['real_churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

# Perform undersampling of the majority class in the training set
rus = RandomUnderSampler(sampling_strategy=sampling_strategy, random_state=random_state)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

# Generate a list of uniformly spaced values for C
C_values = np.random.uniform(0.01, 100, 100)

# Define the hyperparameter search space
param_distributions = {
    'C': C_values,  # List of uniformly spaced values of C
    'solver': ['liblinear', 'saga']
}

# Configure RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=LogisticRegression(random_state=random_state),
    param_distributions=param_distributions,
    n_iter=n_iter,  # Number of combinations to try
    cv=cv_folds,    # Number of folds for cross-validation
    scoring='accuracy',
    random_state=random_state  # Set seed for reproducibility
)

# Train the model
random_search.fit(X_resampled, y_resampled)

# Get the best model
model = random_search.best_estimator_

# Print the best parameters found
print(f"Best parameters found: {random_search.best_params_}")
````

### Churn Model Assessment

````python
# Make predictions
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

# Display the confusion matrix on the test set
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print('====================')
print(pd.DataFrame(conf_matrix, columns=['Predicted 1', 'Predicted 0'], index=['Actual 1', 'Actual 0']))
print()

# Calculate and display performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print('Performance Metrics:')
print('===========================')
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}\n')
````

````python
# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

# Plot the ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'AUC = {roc_auc:.3f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend(loc='lower right')
plt.show()
````
